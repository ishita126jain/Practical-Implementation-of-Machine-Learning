```python
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import numpy as np
from sklearn.cluster import KMeans
```

```python
X = np.array([[1,2],
             [1.5,1.8],
             [5,8],
             [8,8],
             [1,0.4],
             [9,11]])
```

```python
plt.scatter(X[:,0],X[:,1],s =50)
plt.show()
```

```python
clf = KMeans(n_clusters=2)
clf.fit(X)
centoroids = clf.cluster_centers_
labels = clf.labels_

colors =["g.","r.","c.","b.","k.","o."]

for i in range(len(X)):
    plt.plot(X[i][0],X[i][1],colors[labels[i]],markersize=10)
plt.scatter(centoroids[:,0],centoroids[:,1],marker='x',s=50)
plt.show()
```

```python
clf = KMeans(n_clusters=6)
clf.fit(X)
centoroids = clf.cluster_centers_
labels = clf.labels_

colors =10*["g.","r.","c.","b.","k."]

for i in range(len(X)):
    plt.plot(X[i][0],X[i][1],colors[labels[i]],markersize=10)
plt.scatter(centoroids[:,0],centoroids[:,1],marker='x',s=150)
plt.show()
```
